## Bài 2. Số Lớn Thứ Hai
**Input**: Một số nguyên *n > 0* và *n* số nguyên.  
**Output**: Số lớn thứ hai trong các số đã nhập. In ra "Not found" nếu không tìm thấy.  
**Ví dụ**:  
Ví dụ 1:
- Input: n = 4, arr = [5, 4, 1, 12]
- Output: result = 5
- Giải thích: Số lớn nhất là 12, số lớn thứ hai là 5

Ví dụ 2:
- Input: n = 4, arr = [5, 4, 1, 12, 5, 12]
- Output: result = 5
- Giải thích: Có hai số cùng lớn nhất là 12, hai số cùng lớn thứ hai là 5

```python
# get input array
def get_input_arr():
    
    print("Input the number n: ", end='')
    n = int(input())

    arr = []
    for i in range(n):    
        print("Input number {}: ".format(i+1), end='')
        arr += [int(input())]
        
    return arr
```

```python
# METHOD 1: Find biggest, then find second biggest
def find_second_biggest(arr):

    n = len(arr)

    # find biggest number
    biggest = arr[0]
    for el in arr:
        biggest = max(biggest, el)

    # find second biggest number
    second = None
    for el in arr:
        if el == biggest:
            continue
        if second == None:
            second = el
        else:
            second = max(el, second)
            
    return second
```

```python
# METHOD 2: Find biggest and second biggest in one loop
def find_second_biggest(arr):

    n = len(arr)

    # find biggest and second biggest
    biggest = arr[0]
    second = None
    for el in arr:
        if biggest < el:
            second = biggest
            biggest = el
            
    return second
```

```python
arr = get_input_arr()
result = find_second_biggest(arr)
if result == None:
    print("Not found")
else:
    print(result)
```

## Bài 3. Hình Tứ Giác
**Input**: Một số thực *a* là độ dài một cạnh của hình vuông. Hai số thực *b, c > 0* là độ dài hai cạnh kề nhau của một hình chữ nhật.  
**Output**: Chu vi và diện tích của hình vuông và hình chữ nhật.  
**Yêu cầu**: Sử dụng lại class *Rectangle* ở bài tập trên lớp. Tạo thêm class *Square* sử dụng tính kế thừa. Học viên tự chọn class được kế thừa.  
**Ví dụ**:
- Input: a = 4, b = 12, c = 2
- Output:
  + Square Perimeter: 16
  + Square Area: 16
  + Rectangle Perimeter: 28
  + Rectangle Area: 24

```python
# METHOD 1: Square inherits Rectangle (like in math)
class Rectangle:

    def __init__(self, edge1, edge2):
        self.__edge1 = edge1
        self.__edge2 = edge2
        
    def calculate_perimeter(self):
        return (self.__edge1 + self.__edge2) * 2
    
    def calculate_area(self):
        return self.__edge1 * self.__edge2

    
class Square(Rectangle):
    
    def __init__(self, edge):
        Rectangle.__init__(self, edge, edge)
```

```python
# METHOD 2: Rectangle inherits Square (more appropriate in programming)
class Square():
    
    def __init__(self, edge):
        self._edge = edge
        
    def calculate_perimeter(self):
        return self._edge * 4
    
    def calculate_area(self):
        return self._edge ** 2


class Rectangle(Square):
    
    def __init__(self, edge1, edge2):
        Square.__init__(self, edge1)
        self.__edge2 = edge2  # Rectangle has more specific attributes than Square
        
    def calculate_perimeter(self):
        return (self._edge + self.__edge2) * 2
    
    def calculate_area(self):
        return self._edge * self.__edge2
```

```python
# get input
print('Input the number a: ', end='')
edgeA = float(input())
print('Input the number b: ', end='')
edgeB = float(input())
print('Input the number c: ', end='')
edgeC = float(input())

# init objects
square = Square(edgeA)
rectangle = Rectangle(edgeB, edgeC)

# print result
print()
print("Square perimeter: ", square.calculate_perimeter())
print("Square area: ", square.calculate_area())
print("Rectangle perimeter: ", rectangle.calculate_perimeter())
print("Rectangle area: ", rectangle.calculate_area())
```

## Bài 4. Kiểm Tra Số Nguyên
**Input**: Một số thực *a*.  
**Output**: Kiểm tra số thực *a* có phải là số nguyên hay không. In ra "Integer" hoặc "Not integer".  
**Ví dụ**:  
Ví dụ 1:  
- Input: a = 4
- Output: "Integer"

Ví dụ 2:  
- Input: a = 3.1416
- Output: "Not integer"

```python
# METHOD 1: Catch the exception
def is_int(num_str):
    result = True
    try:
        int(num_str)
    except:
        result = False
    return result
```

```python
# METHOD 2: Convert to float first, then to int
def is_int(num_str):
    num_float = float(num_str)
    return num_float == int(num_float)
```

```python
print("Input the number a: ", end='')
if is_int(input()):
    print("Integer")
else:
    print("Not integer")
```
