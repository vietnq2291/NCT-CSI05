## Bài 1. Số Hoàn Hảo
Số hoàn hảo là một số nguyên dương mà tổng các ước nhỏ hơn nó bằng chính nó.  
**Input**: Một số nguyên *n > 0*.  
**Output**: *n* có là số hoàn hảo hay không.  
**Ví dụ**:

Ví dụ 1:  
- Input: n = 6
- Output: 6 is a perfect number
- Giải thích: Các ước của 6 nhỏ hơn chính nó là 1, 2, 3. Ta có 1+2+3 = 6

Ví dụ 2:
- Input: n = 10
- Output: 10 is NOT a perfect number
- Giải thích: Các ước của 10 nhỏ hơn chính nó là 1, 2, 5. Ta có 1+2+5 != 10

```python
def is_perfect(num):

    div_sum = 0
    for i in range(1, (num//2)+1):  # only loop to n/2
        if num % i == 0:
            div_sum += i
            if div_sum > num:
                return False
            
    return div_sum == num


n = 8128
if is_perfect(n):
    print("{} is a perfect number".format(n))
else:
    print("{} is NOT a perfect number".format(n))
```

## Bài 2. Tổng Giai Thừa
**Input**: Một số nguyên *n > 0*.  
**Output**: Kết quả của biểu thức 1!+2!+3!+4!+…+N!.  
**Ví dụ**:  
- Input: n = 3
- Output: result = 9
- Giải thích: 1!+2!+3! = 1+2+6 = 9

```python
# METHOD 1: Two loops, O(n^2)
# METHOD 2: Below, O(n)
def sum_fact(num):
    
    fact = 1
    result = 1
    for i in range(2, num+1):
        fact *= i
        result += fact
        
    return result


sum_fact(5)
```
